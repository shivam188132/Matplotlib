# Scatter Plot 

```python
import matplotlib.pyplot as plt
```

```python
# syntax
# x =[]
# y= []
#  plt.scatter(x,y)
# plt.show()
```

```python
x= [1,2,3,4,5,6,7]
y = [2,3,3,4,5,6,4]
plt.scatter(x,y)
plt.show()
```

```python
day= [1,2,3,4,5,6,7]
no = [2,3,3,4,5,6,4]
plt.scatter(day,no)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

# changing other parameters

```python
day= [1,2,3,4,5,6,7]
no = [2,3,3,4,5,6,4]
# color
plt.scatter(day,no,color="r")


plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# for more differnt colors with 'c' parameter
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
# color
color_x = ['r', 'b', 'y', 'g', 'm', 'b']

plt.scatter(day, no, c=color_x)


plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# changing the size of dots with 's' parameter
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
color_x = ['r', 'b', 'y', 'g', 'm', 'b']

# size

sizes = [40,30,20, 15,10,5]

plt.scatter(day,no, c=color_x, s=sizes)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# changing transpercy using alpha parameter
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
color_x = ['r', 'b', 'y', 'g', 'm', 'b']

# size

sizes = [40,30,20, 15,10,5]

plt.scatter(day,no, c=color_x, s=sizes, alpha=0.4)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# changing the marker parameter
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
color_x = ['r', 'b', 'y', 'g', 'm', 'b']

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=color_x, s=sizes, marker = "*")
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# changing the edge color of the marker
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
color_x = ['r', 'b', 'y', 'g', 'm', 'b']

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=color_x, s=sizes, marker = "*", edgecolor='g')
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# changing the line width of edge parameter
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
color_x = ['r', 'b', 'y', 'g', 'm', 'b']

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=color_x, s=sizes, marker = "*", edgecolor='g', linewidth=2)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# using cmap parameter and passing colors as numbers
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
# color_x = ['r', 'b', 'y', 'g', 'm', 'b']
color_x = [10,49,30,29,56,20]

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=color_x, s=sizes, cmap="BrBG")

plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# for showing a color bar
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
# color_x = ['r', 'b', 'y', 'g', 'm', 'b']
colors = [10,49,30,29,56,20]

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=colors, s=sizes, cmap="BrBG")
# for creating a color bar at right hand side
plt.colorbar()
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# providing a label to colorbar
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
# color_x = ['r', 'b', 'y', 'g', 'm', 'b']
colors = [10,49,30,29,56,20]

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=colors, s=sizes, cmap="BrBG")
# for creating a color bar at right hand side
name = plt.colorbar()
name.set_label("Color Bar", fontsize=15)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```

```python
# for drawing multiple diagaram
```

```python
day= [1,2,3,4,5,6]
no = [2,3,3,4,5,6]
# extra y data
no2 = [4,8,3,1,4,1]
# color_x = ['r', 'b', 'y', 'g', 'm', 'b']
colors = [10,49,30,29,56,20]

# size

sizes = [200,150,120, 115,110,115]

plt.scatter(day,no, c=colors, s=sizes, cmap="BrBG")
plt.scatter(day,no2, color="r",s=200, alpha=0.5)
# for creating a color bar at right hand side
name = plt.colorbar()
name.set_label("Color Bar", fontsize=15)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize= 15)
plt.ylabel("No.", fontsize= 15)
plt.show()
```
