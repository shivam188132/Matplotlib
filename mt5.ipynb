# Stem Plot

```python
# syntax
# x =[]
# y =[]
# plt.stem(x,y)
# plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y)
plt.show()
```

```python
# how can we change line into dotted line
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':')
plt.show()
```

```python
# changing the marker style from circle to +
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', markerfmt='r+')
plt.show()
```

# rounding it again

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', markerfmt='ro')
plt.show()
```

```python
# changng the bottom line to start from 1
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', bottom=1)
plt.show()
```

```python
# changing the color of the base line
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', basefmt='g')
plt.show()
```

```python
# adding the labels
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', markerfmt='ro', basefmt='g' ,label='python')
# for showing label
plt.legend()
plt.show()
```

```python
# changing lines according to their base color 
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', markerfmt='ro', basefmt='g' ,label='python', use_line_collection=False)
plt.legend()
plt.show()
```

```python
# changing the orientation to horizontal
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y, linefmt=':', markerfmt='ro', basefmt='g' ,label='python', use_line_collection=False
        ,orientation='horizontal')
plt.legend()
plt.show()
```
