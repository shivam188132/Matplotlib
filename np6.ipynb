# Box vs Whisker Plot

```python
import matplotlib.pyplot as plt
```

```python
# synatx
# x =[]
# plt.boxplot(x)
# plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python
# changing the shape of box plot using notch property
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch=True)

plt.show()
```

```python
# for flipping the box along horizontal
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert=False)

plt.show()
```

```python
# changing the width of the box
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, widths=0.5)

plt.show()
```

```python
# changinng the label from 1 to something else
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels=['Python'])

plt.show()
```

```python
# for color filling inside the box
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels=['Python'], patch_artist=True)

plt.show()
```

```python
# we can also the mean
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels=['Python'], showmeans=True)

plt.show()
```

```python
# how to show outler {meaning the value greater than what in our x box }
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'])

plt.show()
```

```python
# if we put another value that is consecutive to 120 for 10 then outler is gone.
# see
```

```python
x = [10,20,30,40,50,60,70,120,130]

plt.boxplot(x, labels=['Python'])

plt.show()
```

```python
# we can join our outler using whis parameter
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], whis=2)

plt.show()
```

```python
# changing the color and shape of outler
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, labels=['Python'], sym='+g')

plt.show()
```

```python
# changing the color of the box
```

```python
x = [10,20,30,40,50,60,70,120]
# boxporps

plt.boxplot(x, labels=['Python'], sym='+g', boxprops=dict(color='r'))

plt.show()
```

```python
# we called this T like sturcture as caps we can change its color to 
```

```python
x = [10,20,30,40,50,60,70,120]
# capprops

plt.boxplot(x, labels=['Python'], sym='+g', boxprops=dict(color='g'),capprops=dict(color='r'))

plt.show()
```

```python
# 2
```

```python
x = [10,20,30,40,50,60,70,120]
# capprops

plt.boxplot(x, labels=['Python'], sym='+g', boxprops=dict(color='g'),capprops=dict(color='r'), 
           whiskerprops=dict(color='b'))

plt.show()
```

```python
# printing multiple box plot together
```

```python
x = [10,20,30,40,50,60,70,120]
x1= [10,20,40,50,30,60,90]
# passing it as list inside y so that positions should be according to list 0,1st positions
y= [x,x1]

plt.boxplot(y, labels=['Python', 'c++'], sym='+g', boxprops=dict(color='g'),capprops=dict(color='r'), 
           whiskerprops=dict(color='b'))

plt.show()
```
