```python
import numpy as np
x1 = ['python', 'c', 'c++', 'java']

y1 = [ 85, 70, 60, 82]
z1 = [20,30,40,50]
# plt.xlabel("languages")
# plt.ylabel("rating")
plt.title("Programming Language", fontsize=20)
# c =("y", "b", "m","g")
# plt.bar(x1,y1, width=0.4, color=c, align="edge", edgecolor="r", linewidth=5, linestyle=':', alpha=0.4)

width=0.2
p1 =np.arange(len(x1))
print(p1)
# for side shifting of 2nd graph
p2 =[ j+ width for j in p1]
plt.bar(p1,y1, width, color='r', align="center",label='popularity')
plt.bar(p2,z1, width, color='y', align="center",label='popularity1')

# for printting all the x1 in bottom
plt.xticks(p1+width/2, x1, rotation=10)

plt.legend()
plt.show()

```

# for getting horizontal bar graphs

```python
import numpy as np
x1 = ['python', 'c', 'c++', 'java']

y1 = [ 85, 70, 60, 82]
z1 = [20,30,40,50]

plt.title("Programming Language", fontsize=20)

width=0.2
p1 =np.arange(len(x1))
print(p1)

p2 =[ j+ width for j in p1]
# use barh insted of bar
plt.barh(p1,y1, width, color='r', align="center",label='popularity')
plt.barh(p2,z1, width, color='y', align="center",label='popularity1')

# yticks
plt.yticks(p1+width/2, x1, rotation=10)

plt.legend()
plt.show()

```
