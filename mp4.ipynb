# Pie Plot

```python
import matplotlib.pyplot as plt
```

```python
# synatx
# x =[]
# plt.pie()
# plt.show()
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']

plt.pie(x)
plt.show()
```

```python
# label parmeter
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']

plt.pie(x, labels=y)
plt.show()
```

```python
# exploding some parts outside the chart to look more attractive
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']

# explode
ex = [0.3, 0.0,0.2,0.0]

plt.pie(x, labels=y, explode=ex)
plt.show()
```

```python
# changing colors
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']

# colors
c = ['r', 'b', 'g', 'y' ]

plt.pie(x, labels=y, colors=c)
plt.show()
```

```python
# showing percentage in the graph using autopct parameter
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']

# colors
c = ['r', 'b', 'g', 'y' ]

plt.pie(x, labels=y, colors=c, autopct ='%0.1f%%')
plt.show()
```

```python
# for 2 decimal places
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']

# colors
# c = ['r', 'b', 'g', 'y' ]

plt.pie(x, labels=y, autopct ='%0.2f%%')
plt.show()
```

```python
# creating shadows
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']


plt.pie(x, labels=y, autopct ='%0.2f%%', shadow=True)
plt.show()
```

```python
# changing radius of pie chart
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']


plt.pie(x, labels=y, autopct ='%0.2f%%', radius=0.69)
plt.show()
```

```python
# adding distance to labels
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']



plt.pie(x, labels=y, autopct ='%0.2f%%', labeldistance=1.2)
plt.show()
```

```python
# we can see the circle is getting started fromm 0 degrees from python
# lets make it start from 90 degrees
```

# rotating circle

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']



plt.pie(x, labels=y, autopct ='%0.2f%%', startangle=90 )
plt.show()
```

```python
# changing font size
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']



plt.pie(x, labels=y, autopct ='%0.2f%%', textprops={'fontsize':15} )
plt.show()
```

```python
# rotating the diagram in clockwise directions
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']



plt.pie(x, labels=y, autopct ='%0.2f%%', counterclock=False)
plt.show()
```

```python
# increasing linewidth 
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']



plt.pie(x, labels=y, autopct ='%0.2f%%',shadow=True, counterclock=False, wedgeprops={"linewidth":10})
plt.show()
```

```python
# changing shadow color
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']
ex = [0.3, 0.0,0.0,0.0]
# edgecolor parameter is used here

plt.pie(x, labels=y, autopct ='%0.2f%%',shadow=True,explode=ex, counterclock=False, wedgeprops={"linewidth":5, 'edgecolor': 'm'})
plt.show()
```

```python
# changing the properties of center
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']
ex = [0.3, 0.0,0.0,0.0]
# center parameter is used

plt.pie(x, labels=y, autopct ='%0.2f%%',shadow=True,explode=ex, counterclock=False,
        wedgeprops={"linewidth":5, 'edgecolor': 'm'},center=(3,6))
plt.show()
```

```python
# rotating labels using rotatelabels parameter
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']
ex = [0.3, 0.0,0.0,0.0]
# rotatelabels parameter is used

plt.pie(x, labels=y, autopct ='%0.2f%%',shadow=True,explode=ex, 
        counterclock=False, wedgeprops={"linewidth":5, 'edgecolor': 'm'},
        rotatelabels=True)
plt.show()
```

```python
# adding a title at the top
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']
ex = [0.3, 0.0,0.0,0.0]


plt.pie(x, labels=y, autopct ='%0.2f%%',shadow=True,explode=ex, 
        counterclock=False, wedgeprops={"linewidth":5, 'edgecolor': 'm'},
        rotatelabels=False)
# 
plt.title('WsCube Tech')
plt.show()
```

```python
# adding legends
```

```python
x=[10,20,30,40]
y = ['c', 'c++', 'java', 'python']
ex = [0.3, 0.0,0.0,0.0]


plt.pie(x, labels=y, autopct ='%0.2f%%',shadow=True,explode=ex, 
        counterclock=False, wedgeprops={"linewidth":5, 'edgecolor': 'm'},
        rotatelabels=False)

plt.title('WsCube Tech')
# changing location with loc parameter inside legend
plt.legend(loc=2)

plt.show()
```
